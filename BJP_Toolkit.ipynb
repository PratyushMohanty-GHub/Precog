```python
df = pd.read_csv(io.StringIO(data['BJP_toolkit - Sheet1.csv'].decode('utf-8')))
display(df)
```

```python
print(df['Tweets'])
print("Size of df: ", len(df))
```

```python
# Remove extra spaces from the tweets
import re

for ind in df.index:
  df["Tweets"][ind] = str(re.sub(' +', ' ', df["Tweets"][ind]))

display(df)
```

```python
# !pip install langid

import langid
# print(langid.classify("जब शरणार्थी भारत आते हैं, तो यहां उन्हें शिक्षा, स्वास्थ्य और दूसरी मूलभूत सुविधाएं नहीं मिल सकती। यह पूर्ण रूप से मानवाधिकारों का हनन है। ये लोग भारतीय नागरिक को मिलने वाली सुविधाओं से वंचित रहते आए हैं. #IndiaSupportsCAA"))
# print(langid.classify("On September 26, 1947, Mahatma Gandhi openly said in a meeting that ""Hindus and Sikhs living in Pakistan can come to India from all angles, if they do not want to reside there. In that case, it is the first duty of the Government of India to give them jobs and make their lives normal. #IndiaSupportsCAA"))

for ind in df.index:
  df.loc[ind, ['Language']] = langid.classify(df['Tweets'][ind])[0]

display(df)
```

```python
# Finding mentions

for ind in df.index:
  mentionList = re.findall("[@]\w+", str(df['Tweets'][ind]))
  if (len(mentionList) > 0):
    df.loc[ind, ['Mentions']] = mentionList;
  else:
    df.loc[ind, ['Mentions']] = "NULL"

display(df)
```

```python
# Finding #Hashtags

for ind in df.index:
  hashtagList = re.findall("[#]\w+", str(df['Tweets'][ind]))
  if (len(hashtagList) > 0):
    df.loc[ind, ['Hashtags']] = hashtagList;
  else:
    df.loc[ind, ['Hashtags']] = "NULL"
display(df)
```
